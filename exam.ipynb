```python
# 🧠 DSA 2040A – Data Warehousing & Mining  
## Mid-Semester Exam (ETL Project)  
### Phase 2 – Transformation

"""Objectives:
1. Load validated datasets from the extraction phase.  
2. Merge raw and incremental data.  
3. Clean and standardize fields.  
4. Handle missing, invalid, and inconsistent records.  
5. Generate the final transformed dataset for loading.
"""

```

```python
raw_validated = pd.read_csv("C:\\Users\\USER\\Desktop\\Data Warehousing\\midsem\\validated_raw.csv")
inc_validated = pd.read_csv("C:\\Users\\USER\\Desktop\\Data Warehousing\\midsem\\validated_incremental.csv")

print(f"Raw validated rows: {len(raw_validated)}")
print(f"Incremental validated rows: {len(inc_validated)}")
```

```python

# MERGE RAW AND INCREMENTAL DATASETS
# ---------------------------------------------
combined_df = pd.concat([raw_validated, inc_validated], ignore_index=True)
combined_df.drop_duplicates(inplace=True)

print(f"Combined dataset rows after merge: {len(combined_df)}")
display(combined_df.head())

#### 💬 Observation

#Merged both datasets into a single dataframe for unified cleaning and transformation.  
#Duplicate records across the two datasets were removed to ensure unique transactions.

```

```python
# HANDLE MISSING VALUES
# Check before handling
print("Missing values before handling:")
display(combined_df.isnull().sum())

# Drop records missing essential fields like Invoice or StockCode
combined_df = combined_df.dropna(subset=['Invoice', 'StockCode'])

# Fill missing descriptions with "Unknown Product"
combined_df['Description'] = combined_df['Description'].fillna('Unknown Product')

# Fill missing customer IDs with placeholder 'Guest'
combined_df['Customer ID'] = combined_df['Customer ID'].fillna('Guest')

print("\nMissing values after handling:")
display(combined_df.isnull().sum())

""" Observation
- Dropped rows missing essential identifiers (`Invoice`, `StockCode`).  
- Filled missing `Description` entries with “Unknown Product.”  
- Replaced missing `Customer ID` values with “Guest” to maintain data consistency.  
All critical fields are now complete with no null values remaining.
"""
```

```python
# ---------------------------------------------
# REMOVE INVALID QUANTITIES AND PRICES
# ---------------------------------------------
invalid_rows = combined_df[(combined_df['Quantity'] <= 0) | (combined_df['Price'] <= 0)]
print(f"Invalid records found: {len(invalid_rows)}")

# Keep only valid rows
combined_df = combined_df[(combined_df['Quantity'] > 0) & (combined_df['Price'] > 0)]

print(f"Remaining rows after cleaning: {len(combined_df)}")
#### 💬 Observation
"""- Removed invalid records with non-positive `Quantity` or `Price`.
- Ensured all remaining records have valid transaction details.
"""
```

```python
# STANDARDIZE DATA TYPES
# ---------------------------------------------

# Convert InvoiceDate to datetime
combined_df['InvoiceDate'] = pd.to_datetime(combined_df['InvoiceDate'], errors='coerce')

# Ensure Quantity and Price are numeric
combined_df['Quantity'] = pd.to_numeric(combined_df['Quantity'], errors='coerce')
combined_df['Price'] = pd.to_numeric(combined_df['Price'], errors='coerce')

# Drop any remaining invalid datetimes
combined_df = combined_df.dropna(subset=['InvoiceDate'])

print("✅ Data types standardized successfully.")
combined_df.info()

# Observation
# Ensured all key fields have appropriate data types for analysis.  
# `InvoiceDate` is now datetime, `Quantity` and `Price` are numeric.
# Dropped any records with invalid dates.
# The transformed dataset is now ready for loading into the data warehouse.
```

```python
# ---------------------------------------------
# ---------------------------------------------
# CREATE DERIVED FIELDS
# ---------------------------------------------
import numpy as np

# Total sales value per record
combined_df['TotalValue'] = combined_df['Quantity'] * combined_df['Price']

# Extract month and year for trend analysis
combined_df['Month'] = combined_df['InvoiceDate'].dt.month
combined_df['Year'] = combined_df['InvoiceDate'].dt.year

# Create a category for customer type
combined_df['CustomerType'] = np.where(combined_df['Customer ID'].astype(str).str.lower() == 'guest', 'Guest', 'Registered')

# Preview sample of new fields
display(combined_df[['Invoice', 'Customer ID', 'CustomerType', 'TotalValue', 'Month', 'Year']].head())

### Observation
"""Added new derived fields:
- **`TotalValue`**: Represents the total amount spent per transaction.  
- **`Month`** and **`Year`**: Extracted for temporal trend analysis.  
- **`CustomerType`**: Differentiates between registered and guest customers.  
These enrichments enhance the dataset’s analytical value.
"""



```
