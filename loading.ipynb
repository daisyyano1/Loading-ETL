```python
# ---------------------------------------------
# LOAD PHASE â€“ DSA 2040A LAB 5
# ---------------------------------------------
import pandas as pd
import sqlite3
import os

# Define folder paths
base_dir = r"C:\Users\USER\Desktop\Data Warehousing\midsem"
transformed_dir = os.path.join(base_dir, "transformed")
loaded_dir = os.path.join(base_dir, "loaded")

# Ensure the loaded folder exists
os.makedirs(loaded_dir, exist_ok=True)

print("âœ… Environment ready. Folders verified.")
```

```python
# Load transformed datasets
full_path = os.path.join(transformed_dir, "transformed_full.csv")
incremental_path = os.path.join(transformed_dir, "transformed_incremental.csv")

df_full = pd.read_csv(full_path)
df_incremental = pd.read_csv(incremental_path)

print(f"Full dataset: {len(df_full)} rows")
print(f"Incremental dataset: {len(df_incremental)} rows")

display(df_full.head())
```

```python
# ---------------------------------------------
# LOAD INTO SQLITE DATABASE
# ---------------------------------------------
db_path = os.path.join(loaded_dir, "full_data.db")
conn = sqlite3.connect(db_path)

# Save both datasets to SQLite
df_full.to_sql('full_data', conn, if_exists='replace', index=False)
df_incremental.to_sql('incremental_data', conn, if_exists='replace', index=False)

print("ðŸ’¾ Data successfully loaded into SQLite database.")
```

```python
# Verify: Read sample rows from each table
print("\nFull Data Sample:")
display(pd.read_sql("SELECT * FROM full_data LIMIT 5;", conn))

print("\nIncremental Data Sample:")
display(pd.read_sql("SELECT * FROM incremental_data LIMIT 5;", conn))

# Confirm record counts
count_full = pd.read_sql("SELECT COUNT(*) as Total FROM full_data;", conn).iloc[0, 0]
count_inc = pd.read_sql("SELECT COUNT(*) as Total FROM incremental_data;", conn).iloc[0, 0]

print(f"\nVerification âœ…\nFull Data Rows: {count_full}\nIncremental Data Rows: {count_inc}")
conn.close()
```
